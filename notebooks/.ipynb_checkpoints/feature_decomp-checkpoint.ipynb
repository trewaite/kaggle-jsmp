```python
import sys
import os
import csv
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
from copy import deepcopy

from sklearn.decomposition import PCA, FastICA

#metrics
from sklearn.metrics import roc_auc_score, accuracy_score

sns.set_style('darkgrid')
pd.set_option('display.float_format', lambda x: '%.3f' % x)

SEED = 26
np.random.seed(SEED)
```

### Read In Data

```python
%%time
# dtype_dict = {}
# for f in range(0,130):
#     dtype_dict[f'feature_{f}'] = 'float32'
train = pd.read_csv('../inputs/train.csv')
```

```python
train.info()
```

```python
%%time
train_fold_ind = pd.read_csv('../preprocessed/train_fold_ind.csv')
```

```python
df = pd.concat([train,train_fold_ind[['fold']]],axis=1)
del train
```

```python
df.head()
```

### Feature Reduction and Correlation

```python
original_features = [f"feature_{f}" for f in range(0,130)]
```

```python
%%time
n_components = 15
pca = PCA(n_components=n_components)
ica = FastICA(n_components=n_components)
df_train_orig = df[df['fold']=='train_fold'][original_features]
df_train_orig = df_train_orig.fillna(0)
```

```python
%%time
pca.fit(df_train_orig.values)
```

```python
%%time
ica.fit(df_train_orig.values)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca_cols = ["pca_"+str(i) for i in range(1,n_components+1)]
ica_cols = ["ica_"+str(i) for i in range(1,n_components+1)]
```

```python
pca_features = pd.DataFrame(pca.transform(df[original_features].fillna(0).values), columns=pca_cols)
```

```python
ica_features = pd.DataFrame(ica.transform(df[original_features].fillna(0).values), columns=pca_cols)
```

```python
df_pca = pd.concat([df, pca_features],axis=1)
```

```python
df_ica = pd.concat([df, ica_features],axis=1)
```

```python
df_pca.head()
```

```python
pca_feature_corr = df_pca.corr()
```

```python
ica_feature_corr = df_ica.corr()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(pca_feature_corr[pca_cols].loc[original_features,:], cmap='viridis')
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(pca_feature_corr[ica_cols].loc[original_features,:], cmap='viridis')
```

```python
plt.figure(figsize=(15,7))
plt.plot(pca_feature_corr['resp'].sort_values(ascending=False)[5:30])
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(15,7))
plt.plot(pca_feature_corr['resp'].sort_values(ascending=True)[5:30])
plt.xticks(rotation=90);
```
